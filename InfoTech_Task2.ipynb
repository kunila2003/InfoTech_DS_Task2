```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
mydata = pd.read_csv("/tmp/tested.csv")
mydata.head()
```

```python
mydata.tail()
```

```python
# Check the dimensions of the dataset
print("Dimensions of the dataset:",mydata.shape)

# Check the data types of each column
print(mydata.dtypes)

# Check for missing values
print("Missing values:")
print(mydata.isnull().sum())

```

```python
mydata.dropna()
```

```python
# Summary statistics for numerical columns
summary_status = mydata.describe()
print("Summary statistics for numerical columns:")
summary_status
```

```python
# Univariate Analysis
# Histogram for Age
plt.figure(figsize=(8, 6))
sns.histplot(mydata['Age'], bins=15, kde=True, color='yellow')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for categorical variables like Sex and Pclass
plt.figure(figsize=(6, 5))
sns.countplot(x='Sex', data=mydata)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Pclass', data=mydata)
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Bivariate Analysis
# Survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=mydata)
plt.title('Survival Rate by Gender')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=mydata)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Scatter plot of Age vs Fare
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=mydata, hue='Survived', palette='Set1')
plt.title('Age vs Fare (Survived)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# Box plot of Age by Passenger Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=mydata)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Pair plot for numerical variables
sns.pairplot(mydata[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', diag_kind='hist')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```
